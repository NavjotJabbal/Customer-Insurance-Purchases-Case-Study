# Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
from matplotlib.colors import ListedColormap
```

# Importing the dataset

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
print(X)
print("____________________________________________________________________________")
print(y)
```

# Splitting the dataset into the Training set and Test set

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
print(X_train)
print("____________________________________________________________________________")
print(y_test)
```

# Feature Scaling

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
print(X_train)
print("_______________________")
print(X_test)
```

# Training the Models on the Training set

```python
log_classifier = LogisticRegression(random_state=0)
log_classifier.fit(X_train, y_train)
knn_classifier = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)
knn_classifier.fit(X_train, y_train)
svm_classifier = SVC(kernel='linear', random_state=0)
svm_classifier.fit(X_train, y_train)
dt_classifier = DecisionTreeClassifier(criterion='entropy', random_state=0)
dt_classifier.fit(X_train, y_train)
rf_classifier = RandomForestClassifier(n_estimators=10, criterion='entropy', random_state=0)
rf_classifier.fit(X_train, y_train)
```

# Predicting a new result

```python
print(f"Logistic Regression: {log_classifier.predict(sc.transform([[30,87000]]))}")
print(f"KNN: {knn_classifier.predict(sc.transform([[30,87000]]))}")
print(f"Support Vector Machine: {svm_classifier.predict(sc.transform([[30,87000]]))}")
print(f"Decision Trees: {dt_classifier.predict(sc.transform([[30,87000]]))}")
print(f"Random Forest (Ensemble learning) Classifiers: {rf_classifier.predict(sc.transform([[30,87000]]))}")
```

# Predicting the Test set results

```python
log_y_pred = log_classifier.predict(X_test)
knn_y_pred = knn_classifier.predict(X_test)
svm_y_pred = svm_classifier.predict(X_test)
dt_y_pred = dt_classifier.predict(X_test)
rf_y_pred = rf_classifier.predict(X_test)

print("Logistic Regression")
print(np.concatenate((log_y_pred.reshape(len(log_y_pred),1),y_test.reshape(len(y_test),1)),1))
print("_________")
print("KNN")
print(np.concatenate((knn_y_pred.reshape(len(knn_y_pred),1),y_test.reshape(len(y_test),1)),1))
print("_________")
print("Support Vector Machine")
print(np.concatenate((svm_y_pred.reshape(len(svm_y_pred),1),y_test.reshape(len(y_test),1)),1))
print("_________")
print("Decision Trees")
print(np.concatenate((dt_y_pred.reshape(len(dt_y_pred),1),y_test.reshape(len(y_test),1)),1))
print("_________")
print("Random Forest (Ensemble learning) Classifiers")
print(np.concatenate((rf_y_pred.reshape(len(rf_y_pred),1),y_test.reshape(len(y_test),1)),1))
```

# Making the Confusion Matrix

```python
log_cm = confusion_matrix(y_test, log_y_pred)
print("Logistic Regression")
print(log_cm)
print(f"Accuracy: {(accuracy_score(y_test, log_y_pred))*100}%")
print("_______________________")
knn_cm = confusion_matrix(y_test, knn_y_pred)
print("KNN")
print(knn_cm)
print(f"Accuracy: {(accuracy_score(y_test, knn_y_pred))*100}%")
print("_______________________")
svm_cm = confusion_matrix(y_test, svm_y_pred)
print("Support Vector Machine")
print(svm_cm)
print(f"Accuracy: {(accuracy_score(y_test, svm_y_pred))*100}%")
print("_______________________")
dt_cm = confusion_matrix(y_test, dt_y_pred)
print("Decision Trees")
print(dt_cm)
print(f"Accuracy: {(accuracy_score(y_test, dt_y_pred))*100}%")
print("_______________________")
rf_cm = confusion_matrix(y_test, rf_y_pred)
print("Random Forest (Ensemble learning) Classifiers")
print(rf_cm)
print(f"Accuracy: {(accuracy_score(y_test, rf_y_pred))*100}%")
```
